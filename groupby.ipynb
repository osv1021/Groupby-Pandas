## Groupby

```python
df.groupby('Country')['TotalInvoice'].sum().sort_values()
```

```python
df.groupby(df['Country'])['TotalInvoice'].sum().sort_values(ascending=False)
```

```python
df['Description'].value_counts()
```

```python
#Con el reset_index se convierte en df
df.groupby('Description')['TotalInvoice'].sum().reset_index()
```

```python
#Con doble corchete también
df.groupby(df['Description'])[['TotalInvoice']].sum()
```

```python
df.groupby('Description')[['UnitPrice', 'Quantity','TotalInvoice']].mean().round(2)
```

```python
df.groupby(['Country','Description'])[['TotalInvoice']].sum().round(2).reset_index()
```

```python
df.groupby('Country')[['Quantity','TotalInvoice']].agg(['mean','sum']).round(2).reset_index()
```

```python
#Si no queremos aplicar las dos a cada columna
columnas_metodos={
    'Quantity':'mean',
    'TotalInvoice':'sum'
}
```

```python
df.groupby('Country')[['Quantity','TotalInvoice']].agg(columnas_metodos).round(2).reset_index()
```

```python
rename_columnas={
    'Quantity':'Quantity_mean',
    'TotalInvoice':'TotalInvoice_sum'
}
```

```python
#Para que se vea mejor y sea más explicativo
df.groupby('Country')[['Quantity','TotalInvoice']].agg(columnas_metodos).rename(columns=rename_columnas).round(2).reset_index()
```
